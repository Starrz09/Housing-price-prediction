##  General Strategy

- For categorical columns where missing values logically indicate **absence** of a feature (garage, pool, fireplace, etc.), we use `"None"` as the fill value.
- For numerical columns related to absent features, we use `0` (e.g., `GarageYrBlt`, `MasVnrArea`).
- For `LotFrontage`, we use **neighborhood-based median imputation** since lot frontage is highly dependent on location.
- Minimal missing categorical values (like `Electrical`) are filled using the **most frequent (mode)** category.
- All transformations are applied to training data first. The same logic will later be **applied consistently to the test set** post-model development.

```python
# Data cleaning
# 1. Imputation & Encoding

def impute_and_encode_ames(df):
    df = df.copy()

    # === Step 1: Handle Missing Values ===
    df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
    df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
    df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())

    none_fill_cols = [
        'Alley', 'MasVnrType', 'BsmtQual', 'BsmtCond',
        'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
        'FireplaceQu', 'GarageType', 'GarageFinish',
        'GarageQual', 'GarageCond', 'PoolQC',
        'Fence', 'MiscFeature'
    ]
    df[none_fill_cols] = df[none_fill_cols].fillna('None')

    if 'Electrical' in df.columns:
        df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])

    # === Step 2: Set Categorical Types (Nominal) ===
    # Identify columns that should be treated as nominal categorical *before* any ordinal mapping
    nominal_features = [
        'MSZoning', 'Street', 'Alley', 'LandContour', 'Utilities',
        'LotConfig', 'Neighborhood', 'Condition1', 'Condition2',
        'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
        'Exterior2nd', 'MasVnrType', 'Foundation', 'Heating', 'CentralAir',
        'Electrical', 'SaleType', 'SaleCondition', 'MiscFeature', 'Fence', 'PoolQC',
        'GarageType', 'MSSubClass'
    ]
    for col in nominal_features:
        if col in df.columns:
            df[col] = df[col].astype('object')


    # === Step 3: Ordinal Encoding ===
    qual_cond_map = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'None': 0}
    bsmt_exposure_map = {'Gd': 4, 'Av': 3, 'Mn': 2, 'No': 1, 'None': 0}
    bsmt_fin_map = {'GLQ': 6, 'ALQ': 5, 'BLQ': 4, 'Rec': 3, 'LwQ': 2, 'Unf': 1, 'None': 0}
    functional_map = {'Sal': 1, 'Sev': 2, 'Maj2': 3, 'Maj1': 4, 'Mod': 5,
                      'Min2': 6, 'Min1': 7, 'Typ': 8, 'None': 0}
    garage_finish_map = {'Fin': 3, 'RFn': 2, 'Unf': 1, 'None': 0}
    paved_drive_map = {'Y': 3, 'P': 2, 'N': 1, 'None': 0}
    lot_shape_map = {'Reg': 0, 'IR1': 1, 'IR2': 2, 'IR3': 3}
    land_slope_map = {'Gtl': 1, 'Mod': 2, 'Sev': 3}

    # Include OverallQual and OverallCond here as they are inherently ordinal
    ordinal_mappings = {
        'ExterQual': qual_cond_map,
        'ExterCond': qual_cond_map,
        'BsmtQual': qual_cond_map,
        'BsmtCond': qual_cond_map,
        'HeatingQC': qual_cond_map,
        'KitchenQual': qual_cond_map,
        'FireplaceQu': qual_cond_map,
        'GarageQual': qual_cond_map,
        'GarageCond': qual_cond_map,
        'PoolQC': qual_cond_map,
        'BsmtExposure': bsmt_exposure_map,
        'BsmtFinType1': bsmt_fin_map,
        'BsmtFinType2': bsmt_fin_map,
        'Functional': functional_map,
        'GarageFinish': garage_finish_map,
        'PavedDrive': paved_drive_map,
        'LotShape': lot_shape_map,
        'LandSlope': land_slope_map,
        'OverallQual': {i: i for i in range(1, 11)}, # Map 1-10 to themselves
        'OverallCond': {i: i for i in range(1, 11)}
    }


    for col, mapping in ordinal_mappings.items():
        if col in df.columns:
            df[col] = df[col].map(mapping).fillna(0).astype(int) # Fill NaN with 0 before converting to int

    # Ensure GarageYearbuilt is int
    df['GarageYrBlt'] = df['GarageYrBlt'].astype(int)


    return df
```

```python
# correlation of house quality and ages
# Calculate property age
current_year = eda_df['YrSold'].max()
eda_df['PropertyAge'] = current_year - eda_df['YearBuilt']
from scipy.stats import spearmanr
corr, p_value = spearmanr(eda_df['OverallQual'], eda_df['PropertyAge'])
print(f"Spearman Correlation: {corr}")
print(f"P-value: {p_value}")
```

```python

# Reverse mappings (number ‚Üí label) from encoding
reverse_mappings = {
    'KitchenQual': {5: 'Ex', 4: 'Gd', 3: 'TA', 2: 'Fa', 1: 'Po', 0: 'None'},
    'ExterQual':   {5: 'Ex', 4: 'Gd', 3: 'TA', 2: 'Fa', 1: 'Po', 0: 'None'},
    'BsmtExposure': {4: 'Gd', 3: 'Av', 2: 'Mn', 1: 'No', 0: 'None'},
    'BsmtFinType1': {6: 'GLQ', 5: 'ALQ', 4: 'BLQ', 3: 'Rec', 2: 'LwQ', 1: 'Unf', 0: 'None'},
    'Functional':  {8: 'Typ', 7: 'Min1', 6: 'Min2', 5: 'Mod', 4: 'Maj1', 3: 'Maj2', 2: 'Sev', 1: 'Sal', 0: 'None'},
    'GarageFinish': {3: 'Fin', 2: 'RFn', 1: 'Unf', 0: 'None'},
    'PavedDrive':   {3: 'Y', 2: 'P', 1: 'N', 0: 'None'},
    'LotShape':     {0: 'Reg', 1: 'IR1', 2: 'IR2', 3: 'IR3'},
    'LandSlope':    {1: 'Gtl', 2: 'Mod', 3: 'Sev'}
}

def plot_feature_impact_boxplots(eda_df, features, titles, palettes=None, orders=None):
    """
    Plots a 2-feature impact analysis using boxplots.
    Automatically maps ordinal numeric features back to category labels for plotting.

    Parameters:
        eda_df (DataFrame): Dataset with features and SalePrice.
        features (list): List of 2 column names to plot.
        titles (list): List of titles for the subplots.
        palettes (list): Optional list of color palettes for each subplot.
        orders (list): Optional list of category orderings.
    """
    fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
    fig.suptitle('Feature Impact on Sale Price', fontsize=16, fontweight='bold')

    for i, feature in enumerate(features):
        ax = axes[i]

        # Copy to avoid modifying original data
        plot_data = eda_df.copy()

        # Map numbers back to categories if mapping exists
        if feature in reverse_mappings:
            plot_data[feature] = plot_data[feature].map(reverse_mappings[feature])

        feature_order = orders[i] if orders else None
        palette = palettes[i] if palettes else None

        sns.boxplot(
            data=plot_data, x=feature, y='SalePrice',
            order=feature_order, ax=ax, palette=palette, hue = feature, legend = False
        )

        ax.set_title(titles[i], fontweight='bold')
        ax.set_xlabel(feature, fontweight='bold')
        if i == 0:
            ax.set_ylabel('Sale Price ($)', fontweight='bold')
        else:
            ax.set_ylabel(None)

        ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x/1000:.0f}K'))
        ax.tick_params(axis='x', rotation=30)

    plt.tight_layout()
    return fig
```

```python
# Feature Engineering

def engineer_features(df):
    df = df.copy()

    # Bathrooms
    df['TotalBathrooms'] = (
        df['FullBath'] + 0.5 * df['HalfBath'] +
        df['BsmtFullBath'] + 0.5 * df['BsmtHalfBath']
    )

    # Total SF
    df['TotalSF'] = df['1stFlrSF'] + df['2ndFlrSF'] + df['TotalBsmtSF']

    # House & Remodel Age
    df['HouseAge'] = df['YrSold'] - df['YearBuilt']
    df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']
    df['Remodeled'] = (df['YearRemodAdd'] != df['YearBuilt']).astype(int)

    # Porch
    porch_cols = ['OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'WoodDeckSF']
    if all(c in df.columns for c in porch_cols):
        df['TotalPorchSF'] = df[porch_cols].sum(axis=1)

    # Flags
    for feat, cols in {
        'HasPool': ['PoolArea'],
        'HasGarage': ['GarageArea'],
        'HasBsmt': ['TotalBsmtSF'],
        'HasFireplace': ['Fireplaces']
    }.items():
        if all(c in df.columns for c in cols):
            df[feat] = (df[cols[0]] > 0).astype(int)

    # Quality per SF
    df['QualSF'] = df['OverallQual'] * df['TotalSF']

    # Bathrooms per Bedroom
    df['BathsPerBedroom'] = df['TotalBathrooms'] / (df['BedroomAbvGr'] + 1)

    # Garage Area per Car
    if all(c in df.columns for c in ['GarageArea', 'GarageCars']):
        df['GarageAreaPerCar'] = df['GarageArea'] / (df['GarageCars'] + 1)

    # Neighborhood rank - Only calculate if 'SalePrice' exists (i.e., for training data)
    if 'SalePrice' in df.columns:
        neighborhood_rank = df['Neighborhood'].map(
            df.groupby('Neighborhood')['SalePrice'].mean().rank(ascending=False, method='dense')
        )
        df['Neighborhood_Category'] = (
            pd.qcut(neighborhood_rank, q=3, labels=['High Rank', 'Medium Rank', 'Low Rank'], duplicates='drop')
        ).astype('object')
    else:
        # For test data, we use a placeholder or handle based on training data ranking if available
        # For now, we'll skip this feature for the test set to avoid the error
        pass # Skip creating this column for the test set

    # Condition Category
    df['Combined_condition'] = (df['Condition1'] + '_' + df['Condition2']).astype('object')

    def map_condition(row):
        conds = [row['Condition1'], row['Condition2']]
        if all(c == 'Norm' for c in conds):
            return 'Neutral'
        elif any(c in ['PosN', 'PosA'] for c in conds):
            return 'Positive'
        elif any(c == 'Feedr' for c in conds):
            return 'Moderate_Negative'
        elif any(c in ['Artery', 'RRAn', 'RRAe', 'RRNn', 'RRNe'] for c in conds):
            return 'Strong_Negative'
        else:
            return 'Neutral'

    df['Condition_Category'] = df.apply(map_condition, axis=1).astype('object')

    # Exterior Material Features
    premium_materials = ['Stone', 'ImStucc', 'BrkFace']
    specialty_materials = ['CemntBd', 'Wd Sdng', 'WdShing']
    standard_materials = ['VinylSd', 'MetalSd', 'HdBoard']

    quality_score_map = {
        'Stone': 5, 'ImStucc': 5, 'BrkFace': 4,
        'CemntBd': 4, 'Wd Sdng': 3, 'WdShing': 3,
        'VinylSd': 3, 'MetalSd': 2, 'HdBoard': 2,
        'Plywood': 1, 'AsbShng': 1, 'BrkComm': 1, 'AsphShn': 1, 'None': 0
    }

    durability_map = {
        'Stone': 'High', 'ImStucc': 'High', 'BrkFace': 'High',
        'CemntBd': 'Medium', 'Wd Sdng': 'Medium', 'WdShing': 'Medium',
        'VinylSd': 'Medium', 'MetalSd': 'Medium', 'HdBoard': 'Medium',
        'Plywood': 'Low', 'AsbShng': 'Low', 'BrkComm': 'Low', 'AsphShn': 'Low', 'None': 'Low'
    }

    def categorize_material(row):
        materials = [row['Exterior1st'], row['Exterior2nd']]
        if any(m in premium_materials for m in materials):
            return 'Premium'
        elif any(m in specialty_materials for m in materials):
            return 'Specialty'
        elif any(m in standard_materials for m in materials):
            return 'Standard'
        else:
            return 'Budget'

    def average_quality(row):
        return np.mean([
            quality_score_map.get(row['Exterior1st'], 0),
            quality_score_map.get(row['Exterior2nd'], 0)
        ])

    def material_consistency(row):
        return 'Single_Material' if row['Exterior1st'] == row['Exterior2nd'] else 'Mixed_Materials'

    def categorize_durability(row):
        dur_levels = [
            durability_map.get(row['Exterior1st'], 'Low'),
            durability_map.get(row['Exterior2nd'], 'Low')
        ]
        if 'High' in dur_levels:
            return 'High'
        elif 'Medium' in dur_levels:
            return 'Medium'
        else:
            return 'Low'

    df['Exterior_Material_Category'] = df.apply(categorize_material, axis=1).astype('object')
    df['Exterior_Quality_Score'] = df.apply(average_quality, axis=1)
    df['Material_Consistency'] = df.apply(material_consistency, axis=1).astype('object')
    df['Durability_Category'] = df.apply(categorize_durability, axis=1).astype('object')

    # Sale Type Grouping
    def sale_type_group(row):
        if 'SaleCondition' in row and row['SaleCondition'] == 'Partial':
            return 'New_Construction'
        elif 'SaleType' in row and 'SaleCondition' in row and row['SaleType'] in ['WD', 'New'] and row['SaleCondition'] == 'Normal':
            return 'Standard_Sale'
        elif 'SaleType' in row and row['SaleType'] in ['COD', 'Con', 'ConLw', 'ConLI', 'ConLD']:
            return 'Alternative_Financing'
        else:
            return 'Other'

    df['SaleCategory'] = df.apply(sale_type_group, axis=1).astype('object')

    # Sale date cyclical month encoding
    if 'MoSold' in df.columns:
        df["MoSold_cos"] = np.cos(2 * np.pi * df["MoSold"] / 12)
        df["MoSold_sin"] = np.sin(2 * np.pi * df["MoSold"] / 12)

    #Log transforming the sales data - Only if 'SalePrice' exists
    if 'SalePrice' in df.columns:
        df['SalePrice'] = np.log1p(df['SalePrice'])


    return df
```

```python
# Specifying the ordinal, passthrough, numerical, and categorical columns
ordinal_columns = [
    'ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual',
    'FireplaceQu','GarageQual','GarageCond','PoolQC','BsmtExposure','BsmtFinType1',
    'BsmtFinType2','Functional','GarageFinish','PavedDrive','LotShape','LandSlope',
    'OverallQual','OverallCond'
]

passthrough_columns = ["MoSold_sin", "MoSold_cos"]

# Numerical columns: strictly numeric types, excluding ordinals and passthrough
numerical_columns = [
    col for col in X_train.select_dtypes(include=["number"]).columns
    if col not in ordinal_columns + passthrough_columns
]

# Categorical columns: remaining columns after removing ordinals, passthrough, and numericals
categorical_columns = [
    col for col in X_train.columns
    if col not in ordinal_columns + passthrough_columns + numerical_columns
]

print(f"Numerical columns ({len(numerical_columns)}):", numerical_columns)
print(f"Categorical columns ({len(categorical_columns)}):", categorical_columns)
```

```python
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

# Transformers
ridge_preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_columns),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns),
    ('pass', 'passthrough', ordinal_columns + passthrough_columns)
])

rf_xgb_preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns),
    ('pass', 'passthrough', [col for col in X_train.columns if col not in categorical_columns])
])

# Pipelines
models = {
    'Ridge': Pipeline([
        ('preprocessor', ridge_preprocessor),
        ('model', Ridge())
    ]),
    'RandomForest': Pipeline([
        ('preprocessor', rf_xgb_preprocessor),
        ('model', RandomForestRegressor(random_state=42))
    ]),
    'XGBoost': Pipeline([
        ('preprocessor', rf_xgb_preprocessor),
        ('model', XGBRegressor(random_state=42))
    ]),
    'LightGBM': LGBMRegressor(categorical_feature=categorical_columns, random_state=42),
    'CatBoost': CatBoostRegressor(cat_features=categorical_columns, random_state=42, verbose=0)
}

# Convert categoricals for LightGBM/CatBoost
X_train_lgb_cat = X_train.copy()
for col in categorical_columns:
    X_train_lgb_cat[col] = X_train_lgb_cat[col].astype('category')

# Train & score
for name, model in models.items():
    if name in ['LightGBM', 'CatBoost']:
        model.fit(X_train_lgb_cat, y_train)
    else:
        model.fit(X_train, y_train)
    score = model.score(X_train_lgb_cat if name in ['LightGBM', 'CatBoost'] else X_train, y_train)
    print(f"{name} training R^2 score: {score:.4f}")
```

```python
from sklearn.model_selection import cross_validate


scoring = {
    'r2': 'r2',
    'rmse': 'neg_root_mean_squared_error'
}

for name, model in models.items():
    if name in ['LightGBM', 'CatBoost']:
        X_cv = X_train_lgb_cat
    else:
        X_cv = X_train

    cv_results = cross_validate(model, X_cv, y_train, cv=5, scoring=scoring)

    mean_r2 = cv_results['test_r2'].mean()
    std_r2 = cv_results['test_r2'].std()
    mean_rmse = -cv_results['test_rmse'].mean()
    std_rmse = cv_results['test_rmse'].std()

    print(f"{name} CV -> Mean R^2: {mean_r2:.4f} ¬± {std_r2:.4f} | Mean RMSE: {mean_rmse:.4f} ")
```
